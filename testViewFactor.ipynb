```python
X = np.logspace(-1, 2, 100)
Y = np.array([0.1, 0.2, 0.4, 0.6, 1, 2, 4, 10, 1000])
L = 1
plt.figure()
for y in Y:
    F = abs.getVFAlignedRectangles(X, y, L)
    plt.loglog(X/L, F, label=f"Y/L={y/L}")
plt.grid(True)
plt.xlabel("X/L")
plt.ylabel("F")
plt.legend()
```
