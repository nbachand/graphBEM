```python
%load_ext autoreload
%autoreload 2
%matplotlib widget

import numpy as np
from matplotlib import pyplot as plt
from model import Radiation as rad
```

```python
X = np.logspace(-1, 2, 100)
Y = np.array([0.1, 0.2, 0.4, 0.6, 1, 2, 4, 10, 1000])
L = 1
plt.figure()
for y in Y:
    F = rad.getVFAlignedRectangles(X, y, L)
    plt.loglog(X/L, F, label=f"Y/L={y/L}")
plt.grid(True)
plt.xlabel("X/L")
plt.ylabel("F")
plt.legend()
```
